#### Since the 'status' column is our target column let's check if it needs any feature engineering

```python
df['status'].value_counts()
```

Here we have 4 classes in our 'status' column, so we can reduce these classes into a 2 class binary classification problem.
After researching it was found that the **'ipo'** class states that the start-up has filed for debt, so we can change that class to closed.\
The 'acquiring' class is the start-up that has been recently acquired so we can assume this to be in the operating class.
To summarize we are going to put the **'operating'** and the **'acquired'** classes to be a binary **1** i.e the operating class or **successful**\
while the **'closed'** and the **'ipo'** will be under binary class **0** i.e closed or **failed** start-ups.

```python
#Making a copy if we need it
data= df.copy()
```

```python
# Let's binarize our classes
df['status'].replace({"acquired":1 ,"operating":1, "ipo":0, "closed":0},inplace=True)
```

```python
df['status'].value_counts()
```

After checking the value counts we can see that the we have two classes '1' and '0' now.

Since we don't have many categories in the 'MainCategory' and 'USA_Region' we can convert them into dummies.\
For the states since we have a lot of states, we can convert it into numbers with Label Encoder.

```python
from sklearn.preprocessing import LabelEncoder

labelencoder=LabelEncoder()
df['state_code']=labelencoder.fit_transform(df['state_code'])

df= pd.get_dummies(df,columns=['MainCategory','USA_Region'], drop_first= True)
```
